```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from kmodes.kmodes import KModes
import pickle
from IPython.display import Image
```

```python
# load 'preprocessed_df' which has already been preprocessed after simple EDA

with open('preprocessed_df', 'rb') as f:
    data = pickle.load(f)
```

```python
# clustering with K-Modes algorithm

np.random.seed(11)
km = KModes(n_clusters = 4, init = "Huang", n_init = 3, verbose = 1)
predicted_labels = km.fit_predict(data)
clustered_data = data.copy()
clustered_data['cluster'] = predicted_labels

# save as pickle to do more EDA
clustered_data.to_pickle('clustered_data')
```

> <b><i>clustered_data를 통해 cluster별 자세한 EDA와 전처리를 거쳐 다시 분석에 사용할 생각임<i>

```python
with open('analyzing_df', 'rb') as f:
    analyzing_df = pickle.load(f)
```

```python
plt.figure(figsize=(12,8))
ax = sns.countplot(x = analyzing_df.cluster)
plt.title('Accident Counts per Cluster')
plt.xlabel('Cluster')
plt.ylabel('Counts')

for p in ax.patches:
    counts = int(p.get_height())
    ax.text(p.get_x(), p.get_height()-1500,'{}'.format(counts), color="white", size=20, style = 'italic')
```

<b>
<div>cluster 0: <i>회전운동 중 일어난 사고.즉 차선변경, 우회전 등을 하다가 가볍게 일어난 접촉사고<i></div>
<div>cluster 1: <i>비 오는 날, 젖은 도로에서 일어난 사고. 구조물과 충돌하며 혼자 일으킨 사고가 많았다.<i></div>
<div>cluster 2: <i>뒤에서 콩 받힌 사고. 속도를 줄이며 멈추는 과정에서 이를 미처 인지하지 못해 뒷차량이 와서 받은 사고<i></div>
<div>cluster 3: <i>밤에 일어난 사고. 시야가 좋지 못해 좌회전 도중 혹은 구조물과 충돌한 경우가 많았다<i></div>

> <i>해당 군집화를 뒷받침하는 이유를 살펴보자. 
<div>단, 각 그래프를 볼 때는 절대적인 길이나 count보다는 상대적인 크기를 보는 것이 바람직하다. 예를 들어, cluster 0 사고군의 count가 나머지 모두를 합한 것보다도 더 크므로, 많은 경우 cluster 0이 두드러지게 나타날 것이다. 하지만 전체 사고에서 각 사고군이 차지하는 비중을 감안하고 각 그래프를 해석해야 한다.</div>

<b>
### Cluster 0: 사각지대 군집
><i> <div>lane_code를 보면 right or left turn이 많다.</div> 
<div>movement_code에서는 moving constant가 압도적으로 많다.</div>
<div>또한, 차량의 파손 위치를 보면 앞부분과 옆부분이 많다.</div>
<div>이로 미루어 볼 때, 주행 중이던 차량이 차선변경 및 회전운동을 할 때 백미러 사각지대의 다른 차량을 보지 못해 사고를 낸 경우임을 짐작할 수 있다.</div>

```python
fig,axes = plt.subplots(nrows=3, figsize = (12,18))
for idx, feature in enumerate(['LANE_CODE', 'MOVEMENT_CODE','AREA_DAMAGED_CODE_MAIN']):
    sns.countplot(y = feature, hue = "cluster", data= analyzing_df, ax=axes[idx])
plt.show()
```

<b>
### Cluster 1: 미끌미끌 군집
><i> <div>비 오늘 날씨에, 도로 표면이 젖은 상태에 일어난 사고가 많음을 볼 수 있다.</div> 

```python
fig,axes = plt.subplots(nrows=2, figsize = (12,12))
for idx, feature in enumerate(['WEATHER_CODE','SURF_COND_CODE']):
    sns.countplot(y = feature, hue = "cluster", data= analyzing_df, ax=axes[idx])
plt.show()
```

<b>
### Cluster 2: 뒤에서 콩 군집
><i> <div>어떤 방식으로 충돌(collision)이 발생했는지 살펴보면 same direction rear end, 즉 뒤에서 와서 받힌 경우가 압도적으로 많았다.</div> 
<div>심지어 cluster 2가 자신보다 비중이 훨씬 큰 cluster 0보다도 훨씬 많은 경우를 보였다.</div>
<div>또한, movement를 보면 가만히 있거나 속도를 줄이는 도중 사고가 발생한 경우가 대부분이었다.</div>
<div>이로 미루어 볼 때, cluster 2는 뒤에서 콩 받힌 사고임을 짐작할 수 있다. 즉, 속도를 줄이며 멈추는 과정에서 이를 미처 인지하지 못해 뒷차량이 와서 받은 사고인 것이다. </div>

```python
fig,axes = plt.subplots(nrows=2, figsize = (12,12))
for idx, feature in enumerate(['COLLISION_TYPE_CODE', 'MOVEMENT_CODE']):
    sns.countplot(y = feature, hue = "cluster", data= analyzing_df, ax=axes[idx])
plt.show()
```

<b>
### Cluster 3: 어둑어둑 군집
><i> <div>29592건의 사고 중 3577건으로 가장 적은 비중을 차지하는 cluster 3이다.</div> 
<div>하지만 저녁시간 대에 dark light을 켠 채로 사고가 난 경우가 다른 사고군에 비해 굉장히 많음을 볼 수 있다.</div>
<div>cluster 3는 어두운 저녁, 밤시간대에 시야가 좋지 않아 발생한 사고임을 명백히 확인할 수 있다.</div>

```python
fig,axes = plt.subplots(nrows=2, figsize = (12,12))
for idx, feature in enumerate(['LIGHT_CODE','TIME_PERIOD']):
    sns.countplot(y = feature, hue = "cluster", data= analyzing_df, ax=axes[idx])
plt.show()
```

# Advice to Policy Making

<b>
> '2008년 교통사고비용 추정'(심재익, 유정복)에는 영국 정부가 산출한 '사고 심각도 인명피해사고 1건당 교통사고비용'이 제시되어 있다. 따라서 해당 방법을 적용해 군집별 교통사고비용을 산출하고, 이를 바탕으로 정책 입안에 대한 인사이트를 제시하기로 했다.
<div>No injury를 1로 놓고, Non-incapacitating Injury는 70, Possible Incapacitating Injury는 900, Incapacitating/Disabled Injury는 2500, Fatal Injury는 8000의 가중치를 주었다.</div>

```python
with open('fatality_applied_df', 'rb') as f:
    fatality_df = pickle.load(f)
```

```python
analyzing_df["Fatality_Cost"] = fatality_df["Fatality_Cost"]
analyzing_df.tail()
```

```python
for cluster_no in range(4):
    print("Cost of Cluster {}:".format(cluster_no), analyzing_df[analyzing_df['cluster'] == cluster_no]["Fatality_Cost"].sum())
```

<b>
> 가장 큰 교통사고비용을 발생시키는 cluster 0을 정책입안 대상으로 삼는 것이 적절하다고 판단했다.

<b>
### Cluster 0: 사각지대 군집

><i><div>기술적 한계가 사고발생 원인이라고 생각할 수 있다. </div>
<div>실제로 운전을 하다보면 주변을 잘 살폈음에도 불구하고 사이드 미러의 사각지대에서 갑자기 나타나는 차량 때문에 놀라는 경우가 있다.</div> 
<div>이는 운전자의 부주의, 고령에 따른 순발력 감퇴 등과 별개로 현재 차량에 채택 중인 사이드 미러의 기술적 한계로 판단된다.</div>
<div>따라서 운전자보조시스템(ADAS) 등을 통해 사이드 미러의 사각지대에 차량 유무를 운전자가 감지할 수 있도록 하는 기술 도입이 필요하다. https://www.lifewire.com/decoding-blind-spot-detection-534803</div>
